```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('North America Land Data Assimilation System (NLDAS) Daily Air Temperatures and Heat Index (1979-2011).txt',delimiter='\t',skipfooter=14,engine='python')
```

```python
df
```

```python
df
df.drop(df.columns[df.apply(lambda col: col.isnull().sum() > 3)], axis=1)
```

```python
df['Date'] = pd.to_datetime(df['Month Day, Year'])
```

```python
df.head()
```

```python
p1 = df[['Date','Avg Daily Max Air Temperature (F)']]
p = p1.plot.bar(x='Date', y='Avg Daily Max Air Temperature (F)', label='Plot one', color='Green', title='Air Temperature and Date', figsize=(55,7))
p
```

```python
#My plot shows the average daily max air temperature on each date over a period of over 30 years.
```
