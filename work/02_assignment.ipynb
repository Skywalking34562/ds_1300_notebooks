```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
facts.head(5)
```

```python
#1
facts ['Population_density']=facts['Pop_2020']/facts['Area_land']
facts1 = facts.sort_values(by='Population_density',ascending=False)
facts1.head(5)
```

```python
#2
facts ['area_tot']=facts['Area_water']-facts['Area_land']
facts1 = facts.sort_values(by='area_tot',ascending=False)
facts1.head(5)
```

```python
#3
facts ['pop_increase']=facts['Pop_2020']-facts['Pop_2010']
facts1 = facts.sort_values(by='pop_increase',ascending=False)
facts1.head(5)
```

```python
#4
facts ['State_bird']=facts['Pop_2020']=facts['Area_tot']
facts1 = facts.sort_values(by='pop_increase',ascending=False)
facts1.head(5)
```

```python
#5
facts['State_cap']=facts['City_1']
facts1 = facts.sort_values(by='city_1_pop',ascending=False)
facts1.head(5)
```

```python
facts['percent_drp']=facts['city_1_pop']-facts['city_5_pop']/facts['city_1_pop']
facts1 = facts.sort_values(by='percent_drp')
facts1.head(5)
```
